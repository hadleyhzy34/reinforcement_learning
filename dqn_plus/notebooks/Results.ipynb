```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(n):
    def to_binary(n):
        result = []
        for i in range(3):
            result.append(n % 2)
            n //= 2
        return result
    binary = to_binary(n)
    binary = [True if num else False for num in binary]
    DUEL, DOUBLE, PRIORITIZED = binary
    RAM_ENV_NAME = 'LunarLander-v2'
    data = np.load('./code/lunarlander_results/{}_{}_{}_{}_rewards.npy'.format(RAM_ENV_NAME, DUEL, DOUBLE, PRIORITIZED))
    return data
```

```python
# Vanilla DQN
data = read_data(0)
plt.figure(figsize=(17, 8))
plt.plot(np.mean(data, axis=0))
plt.plot(np.quantile(data, 0.95, axis=0), alpha=0.2)
plt.plot(np.quantile(data, 0.05, axis=0), alpha=0.2)
plt.plot([0, 1000], [200, 200], '--', color='r')
plt.plot([0, 1000], [250, 250], '--', color='b')
plt.ylim(-300, 300)
plt.show()
```
